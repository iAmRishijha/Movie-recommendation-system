```python
import numpy as np
import pandas as pd

```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
df = movies.merge(ratings,on='movieId')
```

```python
df.head()
```

```python
df = df.drop('timestamp',axis = 1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
movie_rating_counts = df.groupby(['movieId', 'title'])['rating'].count().reset_index()
movie_rating_counts = movie_rating_counts.rename(columns={'rating': 'total_ratings'})
movie_rating_counts
```

```python
df = df.merge(movie_rating_counts,on=['movieId','title'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
print(df['rating'].unique())
```

```python
len(df['title'].unique())
```

```python
df['total_ratings'].describe()
```

```python
popular_movies = df[df['total_ratings'] >= 60]
```

```python
movie_features_df =  popular_movies.pivot_table(index='title', columns='userId', values='rating').fillna(0)
movie_features_df
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(movie_features_df)
```

```python
movie_features_df.shape
```

```python
i = np.random.choice(movie_features_df.shape[0])
i = 110
```

```python
dist , indx = knn.kneighbors(movie_features_df.iloc[i,:].values.reshape(1,-1), n_neighbors=6)
```

```python
indx = indx.flatten()
```

```python
for x in indx:
    if x==indx[0]:
        print("Reccomended Movies For : "+ movie_features_df.index[x] )
    else:
        print(movie_features_df.index[x])
```
